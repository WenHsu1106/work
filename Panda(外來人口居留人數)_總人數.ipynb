```python
aa=DF.loc[:,'美國':'其他'].T #篩選標籤從 美國 到 其他， :符號表示該行全部資料(也可以設定範圍例如 0:2)
bb =[]
for index,row in aa.iterrows(): #取出aa當中標籤(index)及對應的資料(row)
      if row.mean()>1000: #平均值大於1000
        bb.append(index) #標籤插入bb list當中
        
#--摺線圖 matplotlib版本 --#        
DF.plot(x='年度', y=bb)
plt.xlabel("Year")
plt.ylabel("Number of people")
plt.show()
```
